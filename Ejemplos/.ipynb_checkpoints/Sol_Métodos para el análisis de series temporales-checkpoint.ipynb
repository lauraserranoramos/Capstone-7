## <font color="#00586D"> 2. Smoothing</font>
<br>

Al trabajar con series temporales, es habitual querer suavizar los valores de la serie, con el objetivo de eliminar ruido y para mostrar mejor la "señal" del proceso subyacente. La técnica más utilizada para esto es la media móvil (del termino en inglés, _moving average_). 

La aplicación de la media móvil da como resultado una nueva serie, creada a partir de la aplicación de la media de una serie de observaciones de la serie original. Normalmente, estos valores se especifican con una **ventana**, que define el número de observaciones a usar para obtener la media. La parte "móvil" hace referencia al hecho de que esta ventana se desliza a lo largo de la serie temporal para ir calculando las medias que se usarán en la nueva serie. 

En `pandas` es muy sencillo el cálculo de estas medias móviles, y, en general, la realización de cálculos sobre ventanas, usando [`df.rolling`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html) que nos permite aplicar diferentes operaciones de agregación sobre ventanas de tiempo. En primer lugar visualicemos la serie de temperaturas para la última década. 

```python
ts = pd.read_csv(
    './data/meteo_retiro.csv', 
    sep=';', 
    parse_dates=['FECHA'], 
    index_col='FECHA', 
    encoding='latin9'
)['TMEDIA']
```

```python
ts['2010':].plot(figsize=(20,5), color='firebrick', ls='-');
```

Como vemos, a simple vista podemos ver que la serie se podría beneficiar de un suavizado. Para ello, vamos a aplicar una ventana de 7 días y la media. 

```python
ts['2010':].plot(figsize=(20,5), color='silver', ls=':');
ts['2010':].rolling(7).mean().plot(figsize=(20,5), color='firebrick', ls='-');
plt.legend(["Original", "Suavizada"])
```

Como se puede ver, lo valores se han suavizado bastante y la señal se aprecia mejor. Aumentando el valor de la ventana podemos obtener incluso series más suavizadas.

```python
ts['2010':].plot(figsize=(20,5), color='silver', ls=':');
ts['2010':].rolling(15).mean().plot(figsize=(20,5), color='firebrick', ls='-');
plt.legend(["Original", "Suavizada"])
```

Veamos más en detalle qué es lo que está haciendo el método rolling anterior. Para ello, vamos a ver, en formato de tabla, tanto la serie original como el resultado de aplicar este suavizado. 

```python
window_size = 7
tb = pd.DataFrame({'y':ts.loc['2010':'2019']})
tb['y_smooth15'] = tb['y'].rolling(window_size).mean()
```

```python
tb.head(10)
```

Como se puede ver en la tabla anterior, los valores de la serie suavizada para las 6 primeras marcas de tiempo son `NaN`, esto es así porque el suavizado con un tamaño de ventana $N$ necesita $N$ valores de la serie original para poderse calcular. Por ello, las 6 primeras marcas aparecen sin valor, aunque, como la serie es bastante larga, no se apreciaba en la visualización. A partir de la marca 7, podéis ver que ya obtenemos un valor, que se corresponde con la media de los 7 valores que se han visto de la serie original. 

```python
tb.loc['2010-01-01':'2010-01-07', 'y'].mean()
```

De esta forma, la ventana de valores se va moviendo, calculando las medias sucesivamente. Por ejemplo, el siguiente valor es:

```python
tb.loc['2010-01-02':'2010-01-08', 'y'].mean()
```

Este tipo de cálculos son muy útiles, no solo de cara al suavizado de ruido en una señal, si no, también, a la hora de crear características para un problema con una clara importancia del tiempo.

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 1</font>

Implementa la función `window_features` que, dada una serie con datos diarios para una serie (por ejemplo, una temperatura), retorna un DataFrame con una columna por cada característica de ventana que se le pida para ese instante. 

La lista de características es una lista de tuplas con dos elementos, la característica (que puede ser mean, sum, o std) y el tamaño de ventana (mirad el ejemplo a continuación). 

```python
def window_features(s: pd.Series, features: list) -> pd.DataFrame:
    """
    Genera características temporales
    """

    df = pd.DataFrame({'y':s.copy()})
    
    ######## Solución
    for feat,window in features:
        name = f'y_{feat}_{window}'
        new_feat = s.rolling(window)
        df[name] = getattr(new_feat, feat)()
    #########
    
    return df
        
```

```python
s = ts.loc['2010':'2019']
window_features(s, [('mean', 3), ('mean', 7), ('sum', 7), ('std', 14)]).tail()
```

Otro método muy utilizado con unas características similares es el suavizado exponencial. Este suavizado no necesita un periodo de inicio como en el caso anterior y es muy sencillo de definir matemáticamente:
$$
S_t =\begin{cases}
 Y_1,                                           & t = 1 \\
 \alpha Y_t + (1 - \alpha) \cdot S_{t-1}, & t > 1
\end{cases}
$$

Esto es, este método utiliza un parámetro $\alpha$ que nos permite regular cuanto se tienen en cuenta valores anteriores a la hora de obtener el valor para el instante $t$, de forma que, a mayor $\alpha$ más se tiene en cuenta los valores más nuevos y a menor $\alpha$, más se tienen en cuenta los valores del pasado. 

Podemos utilizar este suavizado con `pandas` usando [`df.ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html#pandas.DataFrame.ewm). 

```python
tb['y_ewm_mean_15'] = tb['y'].ewm(alpha=0.1).mean()
```

```python
tb.head(10)
```

```python
tb[['y', 'y_ewm_mean_15']].plot(
    figsize=(20,5), 
    color=['silver', 'firebrick'], 
    ls='-'
);
```

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 2</font>

Implementa la función `ewm_features` que, dada una serie con datos diarios para una serie (por ejemplo, una temperatura), retorna un DataFrame con una columna por cada característica de media móvil exponencial que se le pida para ese instante. 

La lista de características de parámetros $\alpha$

```python
def ewm_features(s: pd.Series, alphas: list) -> pd.DataFrame:
    """
    Genera características temporales
    """

    df = pd.DataFrame({'y':s.copy()})
    
    ######## Solución
    for alpha in alphas:
        name = f'y_ewm_{alpha}'
        df[name] = s.ewm(alpha=alpha).mean()
    #########
    
    return df
        
```

```python
s = ts.loc['2010':'2019']
ewm_features(s, [0.3, 0.1, 0.01]).tail()
```

## <font color="#00586D"> 3. Resampling e interpolación</font>
<br>

A veces, nos encontramos con observaciones que tienen una frecuencia que no se corresponde con la necesitamos. A veces es demasiado alta para nuestras necesidades, por ejemplo, cuando tenemos una frecuencia minutal, pero solo nos interesan los valores horarios, y en otros casos, no tienen la frecuencia suficiente, cuando por ejemplo tenemos datos cada 5 minutos pero nos interesarían cada minuto. Para poder resolver esto, normalmente se utiliza la técnica de *resampling* en conjunto con la interpolación. 

El *resampling* consiste en cambiar la frecuencia de los datos. Existen dos tipos:
* Downsampling: Cuando disminuímos la frecuencia (P.e. de segundos a minutos)
* Upsampling: Cuando incrementamos la frecuencia (P.e. de minutos a segundos)


En ambos casos, nos "inventaremos" unos datos nuevos. Para el caso de *downsampling*, tendremos que elegir el estadístico para calcular los nuevos agregados (por ejemplo, la media). Para el caso de *upsampling* tendremos que determinar como obtener las nuevas observaciones usando interpolación. 


### <font color="#00586D"> 2.1  Downsampling </font>

Nuestros datos meteorológicos tienen una frecuencia diaria, pero nos interesaría tenerlos cada dos semanas, agregados por la media. Lo podemos hacer de forma muy sencilla con el método [`df.resample`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html)

```python
ts_res = ts["2020":].resample(rule='2W').mean()
ts_res.head()
```

Veamos gráficamente el resultado. 

```python
ts['2020':].plot(figsize=(20,5), color='silver', ls=':');
ts_res.plot(figsize=(20,5), color='firebrick', ls='-');
plt.legend(["Original", "BiSemanal"])
```

Como podéis observar, la serie ahora tiene valores con frecuencia bisemanal, agregando los valores del periodo cada dos domingos. 

 Veamos un poco más en detalle la operación que se está realizando. 

```python
ts_res = ts["2020":].resample(rule='2W').mean()
ts_res.head()
```

```python
ts["2020":].head(30)
```

Comprobemos manualmente el resultado para 2020-01-19

```python
ts["2020-01-06":"2020-01-19"].mean()
```

Como podéis ver, para el periodo entre "2020-01-06" y "2020-01-19" (ambos incluídos) se ha tomado la media y se ha asignado a la marca "2020-01-19". Esto es, podemos pensar que la operación `resample` mueve los datos hacia la siguiente marca de interés y una vez realizado este movimiento, los agrega para obtener un único valor. 

El parámetro más importante al utilizar el método `resample` es la regla de la agregación (también denominado *offset*) ya que es el parámetro que nos permite elegir la frecuencia de los datos deseada. ### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 3</font>

Implementa la función `quarter_values` que, dada una serie con datos diarios (por ejemplo, temperatura), retorna un dataframe pero con frequencia trimestral, con los datos de la media, así como los valores  y la desviación para cada trimestre. Para cada trimestre, el dato se tiene que dar con la marca de tiempo del día inicial del trimestre.  

```python
def quarter_values(s: pd.Series) -> pd.DataFrame:
    """
    Obtiene valores trimestrales
    """

    df = pd.DataFrame()
    
    ######## Solución
    df['media_trimestral'] = s.resample(rule='QS').mean()
    df['std_trimestral'] = s.resample(rule='QS').std()

    #########
    
    return df
        
```

### <font color="#00586D"> 2.2  Upsampling </font>

A diferencia del _downsampling_ el _upsampling_ requiere usar un método adicional aparte del propio de _resampling_, ya que, a diferencia del primero, estaremos generando huecos para los que no tenemos datos con los que calcular un estadístico. Necesitaremos interpolar esos datos. 

El procedimiento es igual al caso de _downsampling_ al principio, ya que deberemos usar el método `df.resample` para modificar la frecuencia a la que tenemos los datos. Supongamos que ahora necesitamos los datos de temperatura de forma horaria.  

```python
hourly = ts["2020":].resample('H').mean()
hourly
```

Como podéis ver, el método `ts.resample` ha creado `NaN` para los datos de los que no disponemos. Sin embargo, esto no es muy útil como está y, normalmente, querremos rellenar estos huecos. Para ello, podemos usar la función [`ts.interpolate`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html). 

```python
hourly_interpol = hourly.interpolate(method='linear')
hourly_interpol
```

En este caso hemos utilizado la interpolación lineal, pero existen otros métodos que podéis consultar en la documentación. Visualizemos el resultado de esta interpolación. 

```python
hourly["2020-01-01":"2020-01-10"].plot(figsize=(20,5), color='black', marker='o');
hourly_interpol["2020-01-01":"2020-01-10"].plot(figsize=(20,5), color='firebrick', ls='-');
plt.legend(["Original", "BiSemanal"])
```

En el gráfico se muestran los datos diarios (original) en negro y los datos interpolados en rojo. Como se puede ver, se obtienen los datos interpolados linealmente que coméntabamos, realizando interpolación entre los datos de los que disponemos.

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 4</font>

El método interpolate no solo es muy interesante de cara a realizar _upsampling_ si no que, como podréis intuir, es muy útil para rellenar datos perdidos. Implementa la función `fill_by_interpolation` a continuación, que toma como entrada una serie con valores perdidos y devuelve una serie sin valores perdidos, usando interpolación. 

```python
def fill_by_interpolation(s: pd.Series) -> pd.DataFrame:
    """
    Rellena valores perdidos mediante interpolación
    """
    ######### Solución
    return s.interpolate(method='linear')
    
```

```python
fill_by_interpolation(
    hourly_interpol['2020-01-01':'2020-01-02']
).tail(10)
```

## <font color="#00586D"> 4. Shifting </font>
<br>

Por último, vamos a ver el método de _shifting_, que nos permite adelantar o atrasar una serie en un cierto número de marcas de tiempo. Aunque esta operación parece poco útil, la usaremos bastante en el apartado de pronóstico. Asimismo, nos permite, de forma sencilla, convertir un problema de serie temporal en un problema de regresión. 

Para realizar el _shift_ es tan fácil como utilizar la función `ts.shift` de `pandas`. Tomemos la siguiente serie de ejemplo. 

```python
t = ts["2020-01"]
t.head()
```

```python
t.tail()
```

Si realizamos un shift de un periodo, obtendríamos lo siguiente. 

```python
t_shift1 = t.shift(periods=1)
```

```python
t_shift1.head()
```

```python
t_shift1.tail()
```

Como podéis ver, los valores de la serie se han adelantado 1 posición. Podemos hacer lo mismo hacia delante, por ejemplo, 2 posiciones. 

```python
t_shift2m = t.shift(periods=-2)
```

```python
t_shift2m.head()
```

```python
t_shift2m.tail()
```

Es la misma operación pero retrasando la serie. 

Lo interesante de estos  shift es que nos permiten, de una forma muy sencilla, obtener características temporales que nos pueden permitir tratar un problema de pronóstico como un problema de machine learning. Por ejemplo, podemos añadir varias características de media móvil y después usar el shift para que el objetivo sea pronosticar el valor de la serie para el día siguiente (o incluso, varios días hacia delante).
