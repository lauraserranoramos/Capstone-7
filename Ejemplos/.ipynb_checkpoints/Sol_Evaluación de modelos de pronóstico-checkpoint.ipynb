# <font color="#004D7F"> 2. Backtesting </font>

<br>

A parte de las cuestiones que ya conocéis de los módulos de ML, cuando se trabaja con series temporales (y en general, con secuencias) es necesario tener en cuenta el caracter temporal de las secuencias, por lo que en la evaluación de este tipo de modelos no podemos valernos de un _sampling_ convencional para obtener el conjunto de datos de entrenamiento y de test. Este tipo de métodos asumen que no hay ninguna relación entre observaciones, que cada observación es independiente, lo que no es verdad en las series temporales (de hecho, es la relación entre observaciones la que queremos modelar). Debido a esto, no podemos dividir los datos en grupos y, por tanto, deberemos dividir los datos con respecto al orden temporal en el que se observaron los datos. En el pronóstico de series, en los que se conoce como **backtesting**. 

Lo anterior, se traduce en que, para evaluar los modelos de pronóstico lo usual es dividir el dataset siguiendo el orden, dejando, dependiendo del problema, una cierta cantidad para test. Esta cantidad puede ser, por ejemplo, un cierto porcentaje, o la podemos expresar como un periodo de tiempo al final de nuestra serie (por ejemplo, hacer un pronóstico el último año de temperaturas). Esto último es muy sencillo usando `pandas`. 

```python
ts = pd.read_csv(
    './data/meteo_retiro.csv', 
    sep=';', 
    parse_dates=['FECHA'], 
    index_col='FECHA', 
    encoding='latin9'
).loc['2000':, 'TMEDIA'].fillna(method='ffill')

ts.plot(figsize=(20,5), color='firebrick', ls='-');
```

```python
training = ts['2000':'2019']
test = ts['2019':]
```

```python
training.plot(figsize=(20,5), color='firebrick', ls='-');
test.plot(figsize=(20,5), color='silver', ls='-');
plt.legend(['Training', 'Test'])
```

Podemos hacer esto también usando un porcentaje (en este caso, 10%). 

```python
n = len(ts) 
split = 0.1 # % de datos para test
n_training = int((1-split)*n)
training = ts[:n_training]
test = ts[n_training:]

training.plot(figsize=(20,5), color='firebrick', ls='-');
test.plot(figsize=(20,5), color='silver', ls='-');
plt.legend(['Training', 'Test'])
```

Con estas divisiones ya estaríamos en condiciones de evaluar nuestro modelo. En realidad, siguiendo esta filosofía es relativamente sencillo ver como podemos realizar múltiples evaluaciones sobre nuestros modelos. 

Sabiendo como podemos dividir el conjunto de entrenamiento, ya solo quedaría evaluar el modelo. De momento, todavía no habéis visto como entrenar un modelo de pronóstico (en lo que nos centraremos en las siguientes prácticas) pero sí que podéis crear modelos simples con las herramientas que tenéis hasta ahora.

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 1</font>

Implementa la función `last_observation_model` que, dada una serie temporal diaria de entrada y un número n de días, devuelve una serie de $n$ observaciones todas con el último valor observado de la serie, a partir de la serie dada como entrada. 

```python
def last_observation_model(ts: pd.Series, n: int) -> pd.Series:
    """
    Modelo que retorna siempre la última observación de los datos
    """

    ######## Solución

    last_time = ts.index[-1]
    last_value = ts[-1]
    pred_range = pd.date_range(
        start=last_time + datetime.timedelta(days=1), 
        periods=n, 
        freq='D'
    )
    values = np.full(n, last_value)
    prediction = pd.Series(values, index=pred_range, name='y')

    #########
    
    return prediction
        
```

```python
prediction = last_observation_model(training, len(test))
training.plot(figsize=(20,5), color='firebrick', ls='-');
test.plot(figsize=(20,5), color='silver', ls='-');
prediction.plot(figsize=(20,5), color='slateblue', ls='-');
plt.legend(['Training', 'Prediction'])
```

Una vez completado el ejercicio anterior tendréis un modelo, muy sencillo, que podréis evaluar. Las métricas de evaluación ya las conocéis. La mayor parte de problemas de pronóstico son problemas de regresión, por lo que podéis utilizar métricas como el RMSE o el MAE. Veamos por ejemplo como utilizar este último. 

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(test, prediction)
```

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 2</font>

Implementa la función `historic_mean_model` que, dada una serie temporal diaria de entrada y un número n de días, devuelve una serie de $n$ observaciones, en la que para cada observación devolveremos la temperatura media para ese día de los años anteriores. 

```python
def historic_mean_model(ts: pd.Series, n: int) -> pd.Series:
    """
    Modelo que retorna siempre la última observación de los datos
    """

    ######## Solución

    last_time = ts.index[-1]
    data = pd.DataFrame({'month':ts.index.month, 'day':ts.index.day,  'y':ts.values})
    previous = data.groupby(['month', 'day'], as_index=False).mean()
    pred_range = pd.date_range(
        start=last_time + datetime.timedelta(days=1), 
        periods=n, 
        freq='D'
    )
    test_df = pd.DataFrame({'month':pred_range.month, 'day':pred_range.day}, index=pred_range)
    test_df = test_df.rename_axis('index').reset_index() # Índice como columna
    pred_df = test_df.merge(previous, on=['month', 'day'], how='left')[['index', 'y']]
    prediction = pred_df.set_index('index')['y'].sort_index()
    #########
    
    return prediction
        
```

```python
prediction = historic_mean_model(training, len(test))
training.plot(figsize=(20,5), color='firebrick', ls='-');
test.plot(figsize=(20,5), color='silver', ls='-');
prediction.plot(figsize=(20,5), color='slateblue', ls='-');
plt.legend(['Training', 'Prediction'])
print('MAE:',  mean_absolute_error(test, prediction))
```

Resultado esperado:
- MAE: 2.3595

Como podéis ver, la estrategia anterior funciona muy bien. Posiblemente, sea difícil encontrar un modelo solo usando componentes temporales que nos permita mejorar los resultados en este problema, aunque pueden ser muy interesante en otros problemas. En general, el enfoque de extracción de características, como el ejemplo anterior, puede ser muy útil en problemas donde hay características que utilizándolas, la componente temporal de los datos sea menos importante. En este caso la media de las temperaturas para los años anteriores para ese mismo día nos da ya mucha información ya que las temperaturas tienden a ser similares de un año para otro.

# <font color="#004D7F"> 3. Validación cruzada temporal </font>

<br>

Aunque ciertamente quizá el método más común para este tipo de problemas es el anterior, definiendo un conjunto de test para un periodo determinado, es interesante ver como podemos realizar validación cruzada en este tipo de problemas. Hay varias formas de hacerlo (cuando veamos `prophet` veréis otra forma de evaluar adicional a esta). Nosotros veremos aquí como usar `TimeSeriesSplit`, el enfoque recomendado en `scikit-learn`. 

`TimeSeriesSplit` es una variación de la validación cruzada _K-Fold_ en la que las primeras _k_ particiones se usan para entrenamiento y la _k+1_ para test. De esta forma, a diferencia de la validación estándar, los conjuntos de entrenamiento sucesivos son superconjuntos de los anteriores. Asimismo, los datos sobrantes del inicio de la serie se añaden al entrenamiento. Veamos un diagrama explicativo. 

![Validación](img/validacion.png)

Veamos un ejemplo de como utilizarlo:

Para utilizarlo, podemos valernos del siguiente método de `scikit-learn`

```python
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)
for y_train, y_val in tscv.split(training):
    training_split = training[y_train].copy()
    test_split = training[y_val].copy()
    training_split.plot(figsize=(20,5), color='firebrick', ls='-', xlim=(training.index[0], training.index[-1]));
    test_split.plot(figsize=(20,5), color='silver', ls='-',   xlim=(training.index[0], training.index[-1]));
    plt.show()
    print('Train Size: ', training_split.shape[0])
    print('Test Size: ', test_split.shape[0])
    prediction = historic_mean_model(training_split, len(test_split))
    print('Historic model MAE: ', mean_absolute_error(test_split, prediction))
```

Con esta validación podemos ver como se comporta nuestro modelo a medida que aumentamos el conjunto de entrenamiento, y, además, podemos obtener una estimación, mediante la media, de la métrica que queramos evaluar pero, en este caso, sobre varias evaluaciones lo que posiblemente nos de una métrica más robusta.

### <font color="#00586D"> <i class="fa fa-pencil-square-o" aria-hidden="true" style="color:#113D68"></i> Ejercicio 3</font>

Implementa la función `temporal_validation_mae` que tome como parámetro una serie como conjunto de entrenamiento, el modelo a validar, y el número de _splits_ y devuelva el MAE medio del modelo sobre ese conjunto. 

```python
from typing import Callable
def temporal_validation_mae(training: pd.Series, model: Callable, splits: int) -> float:
    # Solution
    tscv = TimeSeriesSplit(n_splits=splits)
    for y_train, y_val in tscv.split(training):
        training_split = training[y_train].copy()
        test_split = training[y_val].copy()
        prediction = model(training_split, len(test_split))
        maes = [mean_absolute_error(test_split, prediction)]
    return np.mean(maes)
```

```python
# Solución
print('MAE (historic_mean_model):', temporal_validation_mae(training, historic_mean_model, 5))
print('MAE (last_observation_model):', temporal_validation_mae(training, last_observation_model, 5))
```

Resultado esperado:
- MAE (historic_mean_model): 2.5856
- MAE (last_observation_model): 10.2109
