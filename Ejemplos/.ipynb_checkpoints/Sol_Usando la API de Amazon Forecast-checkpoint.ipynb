```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import time


def wait_for_it(status_f):
    status = status_f()
    t1 = datetime.datetime.now()
    print(datetime.datetime.now(),'Waiting for status ACTIVE')
    while status != 'ACTIVE':
        time.sleep(10)
        status = status_f()
        print('\t', datetime.datetime.now(), status)
    print(f'Elapsed time: {datetime.datetime.now()-t1}')
```

## <font color="#00586D"> 2. Preparación de datos</font>
<br>

Para esta práctica, la idea es crear un modelo de pronóstico sobre la serie de temperaturas con la que venimos trabajando. No existe ningún tipo de _dataset_ de Amazon Forecast para este caso de uso por lo que usaremos el dominio [`CUSTOM`](https://docs.aws.amazon.com/forecast/latest/dg/custom-domain.html). Asimismo, vamos a usar únicamente el conjunto de tipo _Target_ ya que no tendremos información relacionada en el futuro (Prophet no puede usar un histórico de información relacionada ni información de tipo metadatos). Por tanto, necesitaremos formar un conjunto de datos con las siguientes columnas:
- `item_id`: solo usaremos una estación, Retiro. 
- `timestamp`: que tendrá que tener el formato YYYY-MM-DD
- `target_value`: el valor de la temperatura media. 

En primer lugar vamos a preparar los datos en `pandas` para que sigan este formato. Solo utilizaremos los datos del 2010 al 2015 y dejaremos el 2016 para test. 

```python
data = pd.read_csv(
    './data/meteo_retiro.csv', 
    sep=';', 
    parse_dates=['FECHA'], 
    index_col='FECHA', 
    encoding='latin9'
).loc['2010':'2016', 'TMEDIA']

training = data.loc['2010':'2015']
test = data.loc['2016']

df = training.reset_index()
df['item_id'] = 'Retiro'
df
```

Ahora será necesario guardar el conjunto de datos como csv para poder subirlo al servicio. El conjunto de datos se debe subir sin cabecera ni índice (la cabecera se establecerá más adelante, para lo que tendremos que tener en cuenta el orden).

```python
df[['item_id', 'FECHA', 'TMEDIA']].to_csv('data/target_dataset.csv', header=False, index=False)
```

Una vez preparados los datos, el proceso es el siguiente:

0. Crearemos un bucket en S3
1. Subiremos los datos de entrenamiento a S3
2. Crearemos un grupo de _datasets_ en el que incluiremos este. 

Antes de subir los datos a S3 necesitaréis un bucket con el que podáis trabajar. Nosotros ya hemos creado un bucket, que es el que usaremos, pero vosotros tendréis que crear el vuestro propio. 

```python
bucket_name = 'es.cidaen.mc3.m7-p7'
```

A continuación subimos el conjunto de datos de entrenamiento a S3. 

```python
key = 'p7_target_dataset.csv'

s3.Bucket(bucket_name).Object(key).upload_file("data/target_dataset.csv")
```

Tras esto, vamos a definir una serie de configuraciones para que las llamadas a las API sean más sencillas:
- `DATASET_FREQUENCY`: La frecuencia del conjunto de datos (puede ser H para datos horários, D para datos diarios...). Podéis ver los posibles valores [aquí](https://docs.aws.amazon.com/forecast/latest/dg/API_CreateDataset.html#forecast-CreateDataset-request-DataFrequency). 
- `TIMESTAMP_FORMAT`: El formato de las marcas de tiempo. Solo es posible utilizar dos tipos de formato dependiendo de la frecuencia: `"yyyy-MM-dd"` y `"yyyy-MM-dd hh:mm:ss"`
- `project`: El nombre del proyecto de pronóstico para poder identificar sus componentes fácilmente. No es estrictamente necesario pero ayuda. 
- `datasetName`: El nombre del _dataset_. 
- `datasetGroupName`: El nombre del grupo de _datasets_
- `s3DataPath`: El path donde se encuentra el _dataset_ en S3
- `domain`: El dominio, en nuestro caso, `CUSTOM`.  
- `datasetType`: En nuestro caso, es de tipo `TARGET_TIME_SERIES`

```python
DATASET_FREQUENCY = "D" 
TIMESTAMP_FORMAT = "yyyy-MM-dd"
project = 'p7_awsforecast'
datasetName= project+'_ds'
datasetGroupName= project +'_dsg'
s3DataPath = "s3://"+bucket_name+"/"+key
domain="CUSTOM"
datasetType='TARGET_TIME_SERIES'
```

A parte de esta información, necesitaremos tener un rol para Amazon Forecast con acceso al bucket de S3. En una cuenta normal lo crearíamos directamente pero en el caso de las cuentas educate tendremos que usar el LabRole, que podréis encontrar en el listado de roles de IAM. En nuestro caso es el siguiente:

```python
role_arn = 'arn:aws:iam::448730727533:role/LabRole'
```

A continuación creamos el grupo de _datasets_ donde añadiremos nuestro dataset (nos guardamos el ARN porque lo necesitaremos luego). 

```python
create_dataset_group_response = forecast.create_dataset_group(
    DatasetGroupName=datasetGroupName,
    Domain="CUSTOM",
)
datasetGroupArn = create_dataset_group_response['DatasetGroupArn']
print(datasetGroupArn)
```

Tras esto especificamos el _schema_ de los datos. Debemos especificar el nombre y tipo de los atributos en el orden que aparezca en el conjunto de datos. Al menos los atributos `item_id`, `timestamp` y `target_value` deben estar presentes para el tipo _target_ del dominio `CUSTOM`. 

```python
schema ={
   "Attributes":[
      {
         "AttributeName":"item_id",
         "AttributeType":"string"
      },
      {
         "AttributeName":"timestamp",
         "AttributeType":"timestamp"
      },
      {
         "AttributeName":"target_value",
         "AttributeType":"float"
      }
   ]
}
```

Provisionamos el _dataset_ en el servicio de AWS Forecast (todavía no lo hemos importado en el servicio). 

```python
response=forecast.create_dataset(
                    Domain=domain,
                    DatasetType=datasetType,
                    DatasetName=datasetName,
                    DataFrequency=DATASET_FREQUENCY, 
                    Schema = schema
)
datasetArn = response['DatasetArn']
print(datasetArn)
forecast.describe_dataset(DatasetArn=datasetArn)
```

Tras esto, añadimos el _dataset_ al grupo de _datasets_ que creamos antes. 

```python
forecast.update_dataset_group(DatasetGroupArn=datasetGroupArn, DatasetArns=[datasetArn])
```

A continuación, importaremos el conjunto de datos al servicio de Amazon Forecast. Esto sí que copiará los datos al servicio de Amazon Forecast. 

```python
datasetImportJobName = 'EP_DSIMPORT_JOB_TARGET'
ds_import_job_response=forecast.create_dataset_import_job(
    DatasetImportJobName=datasetImportJobName,
    DatasetArn=datasetArn,
    DataSource= {
        "S3Config" : {
                        "Path":s3DataPath,
                        "RoleArn": role_arn
                    } 
    },
    TimestampFormat=TIMESTAMP_FORMAT
)
ds_import_job_arn=ds_import_job_response['DatasetImportJobArn']
print(ds_import_job_arn)
```

Podéis consultar el estado de la importación a continuación. 

```python
status_f = lambda : forecast.describe_dataset_import_job(DatasetImportJobArn=ds_import_job_arn)['Status']
wait_for_it(status_f)
```

Una vez que la importación haya acabado, ya podéis continuar con la práctica.

## <font color="#00586D"> 3. Entrenamiento del predictor </font>
<br>

En esta sección vamos a crear el predictor. Existen multitud de opciones para su creación, así como decisiones en cuanto las características de las predicciones que haran. Nos vamos a centrar en las configuraciones principales pero podéis ver la multitud de opciones disponibles en a [documentación de la API](https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/forecast.html#ForecastService.Client.create_predictor). 

- `AlgorithmArn`: El tipo de algoritmo a utilizar. Nosotros usaremos el ARN de Prophet. 
- `ForecastHorizon`: El número de puntos a pronosticar a partir del final de la serie proporcionada. Nosotros pronosticaremos 1 año completo. 
- `ForecastTypes`: Una lista de cuantiles a predecir. Lo dejaremos por defecto, lo que nos proporcionara los cuantiles P10, P50 y P90. 
- `InputDataConfig`: Para especificar el _dataset group_ de entrada. 
- `FeaturizationConfig`: Nos permite decidir como se van a procesar las series de entrada. Nosotros no haremos ningún procesado, pero sí que necesitaremos indicar la frecuencia del pronóstico, que es diaria.

Veamos como crear el predictor. 

```python
predictorName= project+'_algo'
predictor_arn = 'arn:aws:forecast:::algorithm/Prophet'
forecastHorizon = 365
featurization_config = {"ForecastFrequency": "D"}
```

```python
create_predictor_response=forecast.create_predictor(
    PredictorName=predictorName, 
    AlgorithmArn=predictor_arn,
    ForecastHorizon=forecastHorizon,
    InputDataConfig= {"DatasetGroupArn": datasetGroupArn},
    FeaturizationConfig=featurization_config
)
predictor_arn=create_predictor_response['PredictorArn']
print(predictor_arn)
```

Para ver el estado de la creación del predictor podéis consultar su estado usando la siguiente celda. Una vez que esté disponible, continuaremos al siguiente paso. 

```python
status_f = lambda : forecast.describe_predictor(PredictorArn=predictor_arn)['Status']
wait_for_it(status_f)
```

## <font color="#00586D"> 4. Creación del pronóstico </font>
<br>

Por último, crearemos el pronóstico propiamente dicho, que es el que se podrá consultar. Cuando creemos el pronóstico, el motor de Amazon Forecast entrenará un modelo y creará el pronóstico con el horizonte que le hayamos especificado. Este horizonte estará disponible para su consulta, como veremos a continuación

```python
forecastName= project+'_alg_forecast'
create_forecast_response=forecast.create_forecast(ForecastName=forecastName,
                                                  PredictorArn=predictor_arn)
forecast_arn = create_forecast_response['ForecastArn']
print(forecast_arn)
```

Podéis consultar el estado de la creación usando la celda a continuación. Cuando esté disponible podemos continuar. 

```python
status_f = lambda : forecast.describe_forecast(ForecastArn=forecast_arn)['Status']
wait_for_it(status_f)
```

Una vez creado, ya podemos consultar el pronóstico con el código a continuación. 

```python
forecastResponse = forecastquery.query_forecast(
    ForecastArn=forecast_arn,
    Filters={"item_id":"Retiro"}
)
```

```python
forecastResponse
```

Como podéis ver, AWS Forecast nos devuelve los tres cuantiles que habíamos pedido (p10, p50 y p90).  Podemos tomar uno de ellos y convertirlo en un _dataframe_. 

```python
aws_forecast = pd.DataFrame(forecastResponse['Forecast']['Predictions']['p50'])
```

```python
aws_forecast['Timestamp'] = pd.to_datetime(aws_forecast['Timestamp'].str.split('T').str[0], format='%Y-%m-%d')
```

```python
pred = aws_forecast.set_index('Timestamp')
pred
```

Con esto podemos evaluarlo. 

```python
from sklearn.metrics import mean_absolute_error

test = test.iloc[0:365] #Para testar solo con el número de días del pronóstico

mean_absolute_error(test.fillna(method='ffill'), pred.fillna(method='ffill'))
```

Podemos visualizar el resultado

```python
data.plot(figsize=(20,5), color='silver', ls='-');
pred['Value'].plot(figsize=(20,5), color='darkgreen', ls='-');
plt.legend(['Data', 'Forecast'])
```
